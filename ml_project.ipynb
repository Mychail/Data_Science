# **My ML Project**

## **Load dataset**

```python
# import dependables
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df.head()
```

## **Data Exploration**

```python
df.info()
```

```python
df.describe()
```

## **Data Preparation**

### **Data Seperation as X and y**

```python
y = df["logS"]
y.head()
```

```python
from re import X
# Feature seperation
X = df.drop("logS", axis=1)
X.head()
```

## **Data Splitting**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
print(X_train.shape), print(X_test.shape), print(y_train.shape), print(y_test.shape)
```

## **Model Building**

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

### **Applying the model to predict**

```python
lr_train_pred = lr.predict(X_train)
lr_test_pred = lr.predict(X_test)
```

### **Evaluate Model Performance**

```python
lr_train_mse = mean_squared_error(y_train, lr_train_pred)
lr_train_r2 = r2_score(y_train, lr_train_pred)
lr_train_mae = mean_absolute_error(y_train, lr_train_pred)

lr_test_mse = mean_squared_error(y_test, lr_test_pred)
lr_test_r2 = r2_score(y_test, lr_test_pred)
lr_test_mae = mean_absolute_error(y_test, lr_test_pred)
```

```python
print("LR MSE (Train): ", lr_train_mse)
print("LR MSE (Test): ", lr_test_mse)
print("LR R2 (Train): ", lr_train_r2)
print("LR R2 (Test): ", lr_test_r2)
print("LR MAE (Train): ", lr_train_mae)
print("LR MAE (Test): ", lr_test_mae)
```

```python
lr_results = pd.DataFrame(["Linear Regression", lr_train_mse, lr_train_r2, lr_train_mae, lr_test_mse, lr_test_r2, lr_test_mae]).transpose()
lr_results.columns = ["Method", "Training MSE", "Training R2", "Training MAE", "Test MSE", "Test R2", "Test MAE"]
```

```python
lr_results
```

## **Random Forest Model Building**

```python
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

### **Applying the model to predict**

```python
rf_train_pred = rf.predict(X_train)
rf_test_pred = rf.predict(X_test)
```

### **Evaluate Model Performance**

```python
rf_train_mse = mean_squared_error(y_train, rf_train_pred)
rf_train_r2 = r2_score(y_train, rf_train_pred)
rf_train_mae = mean_absolute_error(y_train, rf_train_pred)

rf_test_mse = mean_squared_error(y_test, rf_test_pred)
rf_test_r2 = r2_score(y_test, rf_test_pred)
rf_test_mae = mean_absolute_error(y_test, rf_test_pred)
```

```python
print("RF MSE (Train): ", rf_train_mse)
print("RF MSE (Test): ", rf_test_mse)
print("RF R2 (Train): ", rf_train_r2)
print("RF R2 (Test): ", rf_test_r2)
print("RF MAE (Train): ", rf_train_mae)
print("RF MAE (Test): ", rf_test_mae)
```

```python
rf_results = pd.DataFrame(["Random Forest", rf_train_mse, rf_train_r2, rf_train_mae, rf_test_mse, rf_test_r2, rf_test_mae]).transpose()
rf_results.columns = ["Method", "Training MSE", "Training R2", "Training MAE", "Test MSE", "Test R2", "Test MAE"]
rf_results
```

## **Compare Models**

```python
pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
```
